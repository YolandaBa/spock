```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filePath = "/home/yba/spock/training_data/resonant/featuresNorbits10000.0Nout80trio"
fileName = "/trainingdata.csv"
# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv(filePath+fileName)
```

```python
# Count NaN values in each row
nan_count_per_row = df.isna().sum(axis=1)

# Identify the positions (column names) of NaN values in each row
nan_positions_per_row = df.apply(lambda row: row.index[row.isna()].tolist(), axis=1)
```

```python
print(nan_positions_per_row)
```

```python
df[nan_count_per_row!=8]['MEGNO'].isna().values.sum()/df[nan_count_per_row!=8].shape[0]
```

```python
# Count NaN values in each column
nan_count_per_column = df.isna().sum()
print(nan_count_per_column)
# Plot the histogram
nan_count_per_column.plot(kind='bar', figsize=(10, 6))
plt.title('Count of NaN Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['nan_count_per_row'] = df.isna().sum(axis=1)
# Count the occurrence of each unique value in 'nan_count'
nan_count_distribution = df['nan_count_per_row'].value_counts().sort_index()
print(nan_count_distribution)
# Plot the histogram
nan_count_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Count of Rows with Different Numbers of NaN Values')
plt.xlabel('Number of NaN Values')
plt.ylabel('Count of Rows')
plt.tight_layout()
plt.show()
```
